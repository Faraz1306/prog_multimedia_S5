```python
from PIL import Image
image = Image.open('trouve_moi_8_bit_message_(16).png')
x, y = image.size
tab = ""

# Parcourir chaque pixel de l'image
for y in range(y):
    for x in range(x):
        px = image.getpixel((x, y))

        # Ignorer la composante alpha si elle existe (on ne garde que R, G, B)
        r, g, b = px[:3]

        # Extraire uniquement le dernier bit de chaque composante (BGR)
        tab += (bin(b)[-1])  # Composante bleu (B)
        tab += (bin(g)[-1])  # Composante vert (G)
        tab += (bin(r)[-1])  # Composante rouge (R)

# Convertir les bits en texte ASCII
result = ""
for i in range(0, len(tab), 8):
    binary_string = tab[i:i+8]  # Prendre les 8 bits
    if len(binary_string) == 8:  # S'assurer qu'il y a bien 8 bits
        ascii_value = int(binary_string, 2)  # Convertir les 8 bits en entier
        character = chr(ascii_value)  # Convertir l'entier en caractère ASCII
        result += character  # Ajouter le caractère au résultat

print(result)
```
