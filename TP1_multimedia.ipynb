<center> <h2> <b> Question 5 : </b> </h2>

Affichez l'histogramme de couleur de votre image. <br><br>

---

L'histogramme de couleur est une représentation graphique qui indique la répartition des différentes valeurs de couleur dans une image. Il est utile pour analyser la composition chromatique d'une image. Dans un histogramme de couleur, les valeurs de couleur sont généralement réparties sur un axe horizontal, tandis que le nombre d'occurrences de chaque valeur est représenté sur l'axe vertical.

En examinant l'histogramme de couleur, vous pouvez obtenir des informations sur la dominance des couleurs dans l'image, la saturation, le contraste et d'autres caractéristiques chromatiques. Cela peut être particulièrement utile dans le traitement d'images et la vision par ordinateur pour comprendre la nature des couleurs dans une image et prendre des décisions en conséquence.
</center>

````python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Ouvrir l'image originale
im = Image.open("poule3.png")

# Convertir en mode RGB (au cas où l'image ne serait pas déjà en mode RGB)
im = im.convert('RGB')

# Convertir l'image en tableau NumPy pour une manipulation plus simple
image_array = np.array(im)

# Séparer les trois canaux R, G, B
R = image_array[:, :, 0].flatten()  # Canal Rouge
G = image_array[:, :, 1].flatten()  # Canal Vert
B = image_array[:, :, 2].flatten()  # Canal Bleu

# Création de l'histogramme pour les trois canaux dans le même graphique
plt.figure(figsize=(10, 6))

# Histogrammes combinés avec différents canaux superposés
plt.hist(R, bins=256, color='red', alpha=0.5, label='Rouge', histtype='stepfilled')
plt.hist(G, bins=256, color='green', alpha=0.5, label='Vert', histtype='stepfilled')
plt.hist(B, bins=256, color='blue', alpha=0.5, label='Bleu', histtype='stepfilled')

# Limiter l'axe des ordonnées (fréquence) entre 0 et 4000
plt.ylim(0, 4000)

# Ajouter un titre et des légendes
plt.title("Histogramme des Couleurs RVB")
plt.xlabel("Valeur de couleur")
plt.ylabel("Fréquence")
plt.legend()

# Afficher le graphique
plt.show()
````


<center> <h2> <b> Question 6 : </b> </h2>  

Effectuez <u> manuellement </u> un retournement vertical (vertical flip) ainsi qu'un retournement horizontal (horizontal flip) de votre image en déplaçant les pixels un par un de votre image originale vers une copie. Vous pouvez egalement utiliser numpy pour réaliser ce genre de traitement.
Ces retournements seront réalisés successivement pour obtenir deux résultats différents. <br><br>

---

**Explication :**

- L'inversion verticale (Vertical Flip) consiste à retourner l'image autour de son axe vertical, ce qui entraîne une inversion de haut en bas. Par exemple, si une image représentant une personne debout est inversée verticalement, elle apparaîtra tête en bas.

- L'inversion horizontale (Horizontal Flip) consiste à retourner l'image autour de son axe horizontal, ce qui provoque une inversion de gauche à droite. Par exemple, si une image d'une personne faisant face à gauche est inversée horizontalement, elle semblera faire face à droite.

Ces opérations sont couramment utilisées dans le traitement d'images pour diverses applications telles que la création de reflets ou la correction d'orientation. Après avoir effectué ces deux retournements, vous obtiendrez deux résultats distincts pour votre image.


</center>

````python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np


def flip_horizontal(image_path):

    im = Image.open(image_path)
    im = im.convert('RGB')

    width, height = im.size
    im_horizontal = Image.new(mode=im.mode, size=(width, height))

    pix_im = im.load()
    pix_horizontal = im_horizontal.load()

    for x in range(width):
        for y in range(height):

            # Obtenir les valeurs RGB du pixel (x, y)
            r, g, b = pix_im[x, y]

            # Calculer la nouvelle position horizontale (flip)
            new_x = width - 1 - x

            # Mettre à jour le pixel à la nouvelle position
            pix_horizontal[new_x, y] = r, g, b

    return im_horizontal


"""""" """Partie optimisée""" """"""

def flip_horizontal_optimise(image_path):

    im = Image.open(image_path)
    im = im.convert('RGB')

    width, height = im.size
    im_horizontal = Image.new(mode=im.mode, size=(width, height))

    pix_im = im.load()
    pix_horizontal = im_horizontal.load()

    width_half = (width + 1) // 2

    for x in range(width_half):
        for y in range(height):

            # Obtenir les valeurs RGB du pixel (x, y)
            r, g, b = pix_im[x, y]

            # Calculer la nouvelle position horizontale (flip)
            new_x = width - 1 - x

            # Obtenir les valeurs du pixel opposé
            r2, g2, b2 = pix_im[new_x, y]

            # Effectuer l'échange des pixels
            pix_horizontal[new_x, y] = r, g, b
            pix_horizontal[x, y] = r2, g2, b2

    return im_horizontal


# Utilisation de la fonction de base
print("fonction de flip horizontal de base: ")
image_flipped = flip_horizontal("logo_iut_montreuil.png")
display(image_flipped)

print("\n")

# Utilisation de la fonction optimisée
print("fonction de flip horizontal optimisée: ")
image_flipped = flip_horizontal_optimise("logo_iut_montreuil.png")
display(image_flipped)

print("\n")

print(f"valeur de width: {width} valeur de height: {height}")

print("\n")




"""""" """Partie Verticale flip""" """"""
def flip_verticale(image_path):

  im = Image.open(image_path)
  im = im.convert('RGB')

  width, height = im.size
  im_verticale = Image.new(mode=im.mode, size=(width, height))

  pix_im = im.load()
  pix_verticale = im_verticale.load()

  new_y = 0

  for x in range(width):
        for y in range(height):

            # Obtenir les valeurs RGB du pixel (x, y)
            r, g, b = pix_im[x, y]

            # Calculer la nouvelle position verticale (flip)
            new_y = height - 1 - y

            # Mettre à jour le pixel à la nouvelle position
            pix_verticale[x, new_y] = r, g, b

  return im_verticale


print("fonction de flip verticale de base: ")
image_flipped = flip_verticale("logo_iut_montreuil.png")
display(image_flipped)
print("\n")

print("source du code: je l'ai fais moi même en m'appuyant sur ce que j'ai fais dans les questions précédentes")
````

À l'aide des bibliothèques Pillow ou Numpy, effectuez une translation de votre image vers la droite, une autre vers la gauche, une vers le haut et une derniere vers le bas. Pour mieux visualiser les résultats, je vous conseille de paramétrer le taux de translation et de le fixer par défaut à une valeur supérieure ou égale à 100 pixels. <br><br>

---

La translation est une opération de transformation d'image qui consiste à déplacer l'ensemble des pixels de l'image dans une direction donnée, généralement horizontalement (vers la droite ou vers la gauche) ou verticalement (vers le haut ou vers le bas). La distance de déplacement est contrôlée par un paramètre, souvent exprimé en pixels.

Lorsque vous effectuez une translation vers la droite, tous les pixels de l'image sont déplacés vers la droite, créant un effet de glissement de l'image dans cette direction. De même, une translation vers la gauche déplace les pixels vers la gauche. Cette opération est utile pour simuler des déplacements d'objets dans une image ou pour ajuster la composition d'une scène.


````python
from PIL import Image

def translation_image(image_path, c=0, f=0):

  im = Image.open(image_path)
  im = im.convert('RGB')

  a = 1
  b = 0
  c = c
  d = 0
  e = 1
  f = f
  im_translated = im.transform(im.size, Image.AFFINE, (a, b, c, d, e, f))
  return(im_translated)

print("fonction de translation d'image vers la droite de 100: ")
image = translation_image("logo_iut_montreuil.png", -100, 0)
display(image)

print("\n")

print("fonction de translation d'image vers la gauche de 100: ")
image = translation_image("logo_iut_montreuil.png", 100, 0)
display(image)

print("\n")

print("fonction de translation d'image vers le haut de 100: ")
image = translation_image("logo_iut_montreuil.png", 0, 100)
display(image)

print("\n")

print("fonction de translation d'image vers le bas de 100: ")
image = translation_image("logo_iut_montreuil.png", 0, -100)
display(image)

print("\n")

print("Source du code: https://stackoverflow.com/questions/37584977/translate-image-using-pil et chatgpt pour optimiser le code")
````

````python
from PIL import Image

""" "Sans ChatGPT" """

def crop_image(image_path, left = 0, up = 0, right = 0, down = 0 ):

  im = Image.open(image_path)

  box = (left, up, right, down)
  cropped_image = im.crop(box)


  return (cropped_image)


""" """ "Avec ChatGPT" """ """

def crop_image_gpt(image_path, left = 0, up = 0, right=None, down=None ):

  im = Image.open(image_path)
  width, height = im.size

  if right is None:
        right = width
  if down is None:
        down = height

  # Ajuster les coordonnées pour éviter des erreurs
  left = max(0, min(left, width))
  up = max(0, min(up, height))
  right = max(left, min(right, width))  # right doit être supérieur ou égal à left
  down = max(up, min(down, height))    # down doit être supérieur ou égal à up



  box = (left, up, right, down)
  cropped_image = im.crop(box)


  return (cropped_image)

print("fonction de recadrage d'image: ")
image = crop_image("logo_iut_montreuil.png", 120,0,470,300)
display(image)

print("\n")

print("fonction de recadrage d'image avec chatgpt: ")
image = crop_image_gpt("logo_iut_montreuil.png", 125,250,465,750)
display(image)

print("\n")

# il y a des erreurs lorsque l'on prends certaines tailles pour le recadrage mais je n'arrive pas à corriger cette erreur

print("source: https://cloudinary.com/guides/automatic-image-cropping/cropping-images-in-python-with-pillow-and-opencv#pillow \n et https://medium.com/@Coursesteach/what-is-meant-by-cropping-an-image-6dcf7628d37e")
````

<h3> <u> La data augmentation : </u> </h3>

La data augmentation est une technique couramment utilisée dans le domaine de l'apprentissage automatique, en particulier dans la vision par ordinateur, pour améliorer la performance des modèles d'apprentissage profond, tels que les réseaux de neurones convolutifs (CNN). Elle consiste à augmenter la quantité de données d'entraînement en appliquant diverses transformations à des images existantes, créant ainsi de nouvelles instances d'entraînement tout en conservant la même étiquette (classe) associée à l'image d'origine.<br><br>


Les pré-traitements couramment réalisés dans la data augmentation incluent :


* Cropping (Recadrage) <br>
* Horizontal Flip (Retournement Horizontal) <br>
* Vertical Flip (Retournement Vertical) <br>
* Redimensionnement <br>
* Translation <br>
* Rotation <br><br>


L'objectif principal de la data augmentation est d'améliorer la généralisation du modèle en l'entraînant sur des données d'entraînement plus diversifiées et en lui permettant de mieux gérer les variations que l'on pourrait rencontrer dans des images du monde réel. Cela permet au modèle d'apprendre des caractéristiques plus robustes et de mieux se comporter lorsqu'il est confronté à de nouvelles données.

Il est important de noter que la sélection des techniques de data augmentation dépend du domaine d'application spécifique et des caractéristiques des données. En ajustant les transformations appropriées, on peut améliorer la capacité du modèle à généraliser tout en évitant le surajustement (overfitting) en d'autre mot qu'il apprenne nos données par coeur.

<br> <h3> <u> Les réseaux à convolution : </u> </h3>

Les réseaux de neurones à convolution, également appelés CNN (Convolutional Neural Networks), sont une classe de modèles d'apprentissage automatique qui sont utilisés pour traiter des données structurées en grille, comme des images. Les CNN ont révolutionné le domaine de la vision par ordinateur et sont couramment utilisés pour la reconnaissance d'images, la classification d'objets, la détection d'objets, et bien plus encore.

Pensez à un réseau de neurones à convolution comme à un détective qui examine une image. Le détective ne regarde pas l'image dans son ensemble, mais plutôt en petits morceaux ou en régions. Ces petites régions sont appelées "filtres" ou "kernels". Le détective déplace ces filtres sur toute l'image et examine chaque région à la recherche de caractéristiques spécifiques, comme des bords, des textures, ou même des formes plus complexes. <br><br>

### <center> Voici une simplification du calcul de convolution : </center>

1. Imaginez une petite fenêtre rectangulaire (le filtre) que vous faites glisser sur l'image, en commençant par le coin supérieur gauche.

2. À chaque position de la fenêtre, vous multipliez les valeurs des pixels de l'image par les valeurs correspondantes du filtre.

3. Vous additionnez toutes ces valeurs multipliées pour obtenir une seule valeur (c'est une opération de sommation).

4. Cette valeur unique représente à quel point la région de l'image correspond au motif que le filtre recherche.

5. Vous déplacez ensuite la fenêtre vers la droite (ou vers le bas, si vous êtes arrivé à la fin de l'image), et vous répétez le processus pour chaque nouvelle région de l'image.

6. À la fin, vous obtenez une nouvelle image (appelée "carte de caractéristiques") qui met en évidence les zones de l'image où le motif recherché a été détecté. <br><br>

En résumé, les réseaux de neurones à convolution sont des outils puissants pour l'analyse d'images, et ils fonctionnent en examinant de petites parties de l'image à la fois pour détecter des motifs importants. Erosion : L'érosion est une opération de traitement d'image qui réduit les objets blancs sur un fond noir. Elle fonctionne en faisant glisser un filtre (ou noyau) sur l'image et en remplaçant chaque pixel par le minimum des pixels sous le filtre. Cela a pour effet de réduire les zones blanches et d'éliminer les petits bruits. <br><br>

Dilatation : La dilatation est l'opposée de l'érosion. Elle agrandit les objets blancs sur un fond noir en remplaçant chaque pixel par le maximum des pixels sous le filtre. Cela a pour effet d'agrandir les zones blanches et de combler les petits trous. <br><br>

Filtre 3x3 : Le filtre (ou noyau) 3x3 est une matrice de taille 3x3 utilisée pour effectuer les opérations d'érosion et de dilatation. Il est appliqué à chaque pixel de l'image en le remplaçant par une valeur calculée en fonction des pixels voisins dans le voisinage 3x3 autour de lui.
